```python
#!pip install pandas
import pandas as pd

# loading the dataset
pd.read_excel("/content/Data Cleaning Practice.xlsx")
df= pd.read_excel("/content/Data Cleaning Practice.xlsx")

#converting submit_time to datetime
#other columns are in the right data type only this needs conversion
df['submit_time']=pd.to_datetime(df['submit_time'])
#df.info()

#cleaning phone number
df["Phone"].unique()
df["Phone"]=df["Phone"].str.replace("GOTEH", " ")
df.head(20)

#cleaning Fullnames
df['Fullnames']=df['Fullnames'].str.title() #convertig to title case
#splitting fullnames
df['Fullnames'].str.split(pat=" ")
df["first_name"]= df['Fullnames'].str.split(pat=" ").str[0]
df["middle_name"]= df['Fullnames'].str.split(pat=" ").str[1]
df["last_name"]=df['Fullnames'].str.split(pat=" ").str[2]
#removing characters attached with names
df['first_name']=df['first_name'].str.rstrip(",.")
df['middle_name']=df['middle_name'].str.rstrip(",.")
df['last_name']=df['last_name'].str.rstrip(",.")
#changing the column index of first name, middle name, and last name
df.insert(1,"first_name",df.pop("first_name"))
df.insert(2,"middle_name",df.pop("middle_name"))
df.insert(3,"last_name",df.pop("last_name"))
df['first_name']=df['first_name'].str.replace("Comr", "Gogorobari")
df['middle_name']=df['middle_name'].str.replace("Gogorobari", "Lesah")
df['last_name']=df['last_name'].str.replace( "Lesah", "Kormaue")
df.iloc[96]
#filling nan values with empty string to aid concatenation back into full names
df['last_name']=df['last_name'].fillna(" ")
df['middle_name']=df['middle_name'].fillna(" ")
#adding names back into a cleaner full names column
df["full_names"]=df["first_name"]+" "+df["middle_name"]+" "+df["last_name"]
df
#taking the new full names column to the first column
df.insert(0,"full_names",df.pop("full_names"))
df
#removing the former full names column
df.drop(labels="Fullnames",axis=1,inplace=True)
#checking for empty values
df.isnull().any()

#cleaning package column
df['Package'].unique()
#replacing graphics design as graphic design
df['Package']=df['Package'].str.replace("Graphics Design","Graphic Design")
df['Package'].unique()

#cleaning duplicates
df[df.duplicated(subset=["full_names","Phone","LGA","Package"], keep=False)]
df.drop_duplicates(subset=["full_names","Phone","LGA","Package"],inplace=True)
df
df[df.duplicated(subset=["full_names","Phone","LGA"], keep=False)]
df.drop_duplicates(subset=["full_names","Phone","LGA"],inplace=True)
df.head(30)
#renumbering row index
df.reset_index(drop=True,inplace=True)
df.head(30)

#removing missing values
df.isnull().any()
df["Phone"]=df["Phone"].fillna(" ")
df.drop(labels="recaptcha", axis=1,inplace=True)
df["full_names"]=df["full_names"].str.lstrip("`")
df["first_name"]=df["first_name"].str.lstrip("`")
df.sort_values(by="full_names",inplace=True)
df.reset_index(drop=True, inplace=True )
df
#downloading cleaned dataset
!pip install openpyxl
df.to_excel("cleaned_data.xlsx", index=True)
```
